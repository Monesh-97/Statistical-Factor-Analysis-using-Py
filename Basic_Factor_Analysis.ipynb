#FACTOR ANALYSIS:

- **1. Factor Analysis**
Exploratory Factor Analysis (EFA):
Use this if you want to explore the underlying structure of your questionnaire data (especially if the factor structure isn’t firmly established).

Example: Examine whether the items in the English speaking confidence questionnaire naturally group into the 3 hypothesized factors.

Confirmatory Factor Analysis (CFA):
If you have a predefined factor structure (3 factors for confidence, 4 for anxiety, 6 for NLP), CFA lets you test how well the observed data fit this model.

Example: Assess the goodness-of-fit for the NLP student version scale’s 6-factor model.

- **2. Reliability Analysis**
Internal Consistency (e.g., Cronbach’s Alpha):
Evaluate the reliability of each subscale. High Cronbach’s alpha values indicate that the items within a factor consistently measure the same construct.

Example: Compute alpha for each of the 3 factors in the English speaking confidence questionnaire.

- **3. Correlation and Regression Analyses**
Correlation Analysis (Pearson or Spearman):
Examine relationships among factors within and between the questionnaires.

Example: Explore if higher English speaking anxiety scores correlate with lower confidence scores.

Multiple Regression Analysis:
Investigate predictive relationships. For instance, you might predict English speaking confidence using factors from the anxiety questionnaire or even combine factors from both questionnaires.

Example: Determine which anxiety factors are significant predictors of speaking confidence.

Structural Equation Modeling (SEM):
This approach can combine CFA and regression in one framework to test a full theoretical model, including latent variables and their relationships.

Example: Model how NLP-related factors influence both anxiety and confidence as latent constructs.

- **4. Group Comparisons (if applicable)**
t-Tests / ANOVA:
If your study involves comparing groups (e.g., comparing scores between different demographics), you can use t-tests (for two groups) or ANOVA (for more than two groups) on factor scores.

Example: Compare the English speaking confidence scores across different age groups.

Multivariate Analysis of Variance (MANOVA):
When you have several dependent variables (for example, multiple factor scores from one questionnaire), MANOVA allows you to test for group differences on a combined set of outcomes.

Example: Assess group differences across all factors of the NLP student version scale.

MANCOVA:
Similar to MANOVA, but includes covariates that might affect the outcomes.

Example: Compare questionnaire scores across groups while controlling for factors like years of experience.

- **5. Advanced Analyses**
Mediation or Moderation Analysis:
These analyses help explore whether the effect of one variable on another is transmitted through a mediator or depends on a moderator.

Example: Investigate whether the relationship between NLP factors and English speaking confidence is mediated by anxiety levels.

Discriminant Function Analysis:
If you need to classify respondents into categories (for example, high vs. low confidence), this analysis can help identify which factors best discriminate between groups.

Example: Determine which factors most effectively differentiate individuals with high versus low English speaking confidence.

Summary
The above methods provide a broad spectrum of inferential techniques you can use:

For assessing structure and reliability: Use EFA, CFA, and reliability (Cronbach’s alpha) analyses.

For examining relationships: Use correlation, regression, and SEM.

For comparing groups: Use t-tests, ANOVA, MANOVA/MANCOVA, and discriminant analysis.

For testing more complex models: Consider mediation/moderation analyses.

Each approach addresses different research questions, whether you’re interested in measurement validation, predicting outcomes, or comparing groups. The specific inferential statistics you choose will depend on your study design, hypotheses, and data characteristics.

## How to check normality of the data

There are several methods to check if your data is normally distributed. Here are some common approaches:

1. Visual Methods
Histogram: Plotting a histogram gives a quick look at the data distribution.

Q-Q Plot (Quantile-Quantile Plot): This plot compares the quantiles of your data against the quantiles of a normal distribution. If the points roughly follow a straight line, the data are likely normally distributed.

2. Statistical Tests
Shapiro-Wilk Test: This test is sensitive for small to medium sample sizes. A p-value above 0.05 suggests that the data do not deviate significantly from normality.

Kolmogorov-Smirnov Test: Another option, although it's less powerful than Shapiro-Wilk for normality.

D'Agostino's K² Test: This test considers skewness and kurtosis.

```python
# Import required libraries
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import ttest_ind
from semopy import Model
import pingouin as pg
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# -----------------------------
# 1. Data Simulation
# -----------------------------
np.random.seed(42)
N = 300  # number of participants

# Simulate data for three questionnaires
data = pd.DataFrame({
    # English speaking confidence (3 indicators)
    'conf1': np.random.normal(0, 1, N),
    'conf2': np.random.normal(0, 1, N),
    'conf3': np.random.normal(0, 1, N),
    # English speaking anxiety (4 indicators)
    'anx1': np.random.normal(0, 1, N),
    'anx2': np.random.normal(0, 1, N),
    'anx3': np.random.normal(0, 1, N),
    'anx4': np.random.normal(0, 1, N),
    # NLP student scale (6 indicators)
    'nlp1': np.random.normal(0, 1, N),
    'nlp2': np.random.normal(0, 1, N),
    'nlp3': np.random.normal(0, 1, N),
    'nlp4': np.random.normal(0, 1, N),
    'nlp5': np.random.normal(0, 1, N),
    'nlp6': np.random.normal(0, 1, N)
})

# Also simulate a grouping variable for group comparisons (e.g., two groups A and B)
data['group'] = np.random.choice(['A', 'B'], size=N)

# -----------------------------
# 2. Exploratory Factor Analysis (EFA)
# -----------------------------
print("=== Exploratory Factor Analysis (EFA) ===\n")

# For English Speaking Confidence Questionnaire (expecting 3 factors)
conf_items = data[['conf1', 'conf2', 'conf3']]
fa_conf = FactorAnalyzer(n_factors=3, rotation='varimax')
fa_conf.fit(conf_items)
print("Confidence Questionnaire Factor Loadings:")
print(pd.DataFrame(fa_conf.loadings_, index=conf_items.columns))
print("\n")

# For English Speaking Anxiety Questionnaire (expecting 4 factors)
anx_items = data[['anx1', 'anx2', 'anx3', 'anx4']]
fa_anx = FactorAnalyzer(n_factors=4, rotation='varimax')
fa_anx.fit(anx_items)
print("Anxiety Questionnaire Factor Loadings:")
print(pd.DataFrame(fa_anx.loadings_, index=anx_items.columns))
print("\n")

# For NLP Student Scale (expecting 6 factors)
nlp_items = data[['nlp1', 'nlp2', 'nlp3', 'nlp4', 'nlp5', 'nlp6']]
fa_nlp = FactorAnalyzer(n_factors=6, rotation='varimax')
fa_nlp.fit(nlp_items)
print("NLP Student Scale Factor Loadings:")
print(pd.DataFrame(fa_nlp.loadings_, index=nlp_items.columns))
print("\n")

# -----------------------------
# 3. Reliability Analysis (Cronbach's Alpha)
# -----------------------------
# Function to compute Cronbach's alpha from a DataFrame of items
def cronbach_alpha(df):
    items = df.shape[1]
    # Compute the correlation matrix of the items
    corr_matrix = df.corr().values
    # Take the upper triangle (without the diagonal)
    triu_indices = np.triu_indices_from(corr_matrix, k=1)
    avg_corr = np.mean(corr_matrix[triu_indices])
    return (items * avg_corr) / (1 + (items - 1) * avg_corr)

# For demonstration, we compute alpha on the "confidence" items.
alpha_conf = cronbach_alpha(conf_items)
print("Cronbach's alpha for Confidence Questionnaire: {:.3f}".format(alpha_conf))
print("\n")

# -----------------------------
# 4. Correlation and Regression Analysis
# -----------------------------
print("=== Correlation and Regression Analysis ===\n")
# Correlation matrix among confidence and anxiety indicators
corr_matrix = data[['conf1', 'anx1', 'anx2', 'anx3', 'anx4']].corr()
print("Correlation matrix between confidence and anxiety items:")
print(corr_matrix)
print("\n")

# Multiple Regression Analysis:
# Predict one of the confidence items (e.g., conf1) using all anxiety items.
X = data[['anx1', 'anx2', 'anx3', 'anx4']]
X = sm.add_constant(X)  # add constant term for intercept
y = data['conf1']
reg_model = sm.OLS(y, X).fit()
print("Regression Summary (Predicting conf1 using anxiety indicators):")
print(reg_model.summary())
print("\n")

# -----------------------------
# 5. Structural Equation Modeling (SEM) / Confirmatory Factor Analysis (CFA)
# -----------------------------
print("=== Structural Equation Modeling (SEM) / CFA ===\n")
# In this example, we specify a simple measurement model with latent variables:
#   - Confidence measured by conf1, conf2, conf3
#   - Anxiety measured by anx1, anx2, anx3, anx4
#   - NLP measured by nlp1,...,nlp6
# And a structural model where:
#   - Confidence is predicted by Anxiety and NLP
#   - Anxiety is predicted by NLP

model_desc = """
# Measurement model
Confidence =~ conf1 + conf2 + conf3
Anxiety =~ anx1 + anx2 + anx3 + anx4
NLP =~ nlp1 + nlp2 + nlp3 + nlp4 + nlp5 + nlp6

# Structural model
Confidence ~ Anxiety + NLP
Anxiety ~ NLP
"""

# Create and fit the SEM model
sem_model = Model(model_desc)
res = sem_model.fit(data)
print("SEM Model Fit Statistics:")
print(res)
print("\nModel Parameter Estimates:")
print(sem_model.inspect())
print("\n")

# -----------------------------
# 6. Group Comparisons: t-Test and ANOVA
# -----------------------------
print("=== Group Comparisons ===\n")
# Independent samples t-test: Compare conf1 scores between groups A and B
group_A = data[data['group'] == 'A']['conf1']
group_B = data[data['group'] == 'B']['conf1']
t_stat, p_val = ttest_ind(group_A, group_B)
print("t-Test for conf1 between groups A and B:")
print("t-statistic = {:.3f}, p-value = {:.3f}".format(t_stat, p_val))
print("\n")

# ANOVA: Compare conf1 scores between groups using linear model
anova_model = smf.ols('conf1 ~ C(group)', data=data).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)
print("ANOVA for conf1 by group:")
print(anova_table)
print("\n")

# -----------------------------
# 7. Mediation Analysis
# -----------------------------
print("=== Mediation Analysis ===\n")
# For demonstration, test if the effect of an NLP indicator (nlp1) on confidence (conf1) is mediated by an anxiety indicator (anx1).
mediation_results = pg.mediation_analysis(data=data, x='nlp1', m='anx1', y='conf1')
print(mediation_results)
print("\n")

# -----------------------------
# 8. Discriminant Function Analysis (DFA) using LDA
# -----------------------------
print("=== Discriminant Function Analysis (LDA) ===\n")
# Use Linear Discriminant Analysis to see which indicators help discriminate between groups.
lda_features = data[['conf1', 'anx1', 'nlp1']]
lda = LinearDiscriminantAnalysis()
lda.fit(lda_features, data['group'])
print("LDA Coefficients:")
print(lda.coef_)
```

# Non-Parametric Test

Non-parametric tests are useful when your data do not meet the assumptions required for parametric tests (e.g., normality). Below are some common non-parametric tests along with example Python code using SciPy:

1. Two-Group Comparisons
Mann-Whitney U Test:
Used to compare differences between two independent groups when the dependent variable is ordinal or not normally distributed.

```python
from scipy.stats import mannwhitneyu

# Example data for two independent groups
group1 = [1.2, 2.3, 2.1, 3.4, 2.8]
group2 = [2.0, 3.1, 1.8, 2.9, 3.3]

stat, p = mannwhitneyu(group1, group2)
print("Mann-Whitney U Test:")
print("Statistic:", stat, "p-value:", p)
```

**Wilcoxon Rank-Sum Test:**
Essentially equivalent to the Mann-Whitney U test (with minor differences in implementation), used for comparing two independent samples.

2. Paired Comparisons
Wilcoxon Signed-Rank Test:
Used for comparing two related or paired samples. For example, comparing scores before and after an intervention.

```python
from scipy.stats import wilcoxon

# Example paired data: before and after scores
before = [1.2, 2.3, 2.1, 3.4, 2.8]
after  = [1.0, 2.5, 2.0, 3.6, 3.0]

stat, p = wilcoxon(before, after)
print("Wilcoxon Signed-Rank Test:")
print("Statistic:", stat, "p-value:", p)
```

**3. Multiple Group Comparisons Kruskal-Wallis H Test:**
An extension of the Mann-Whitney U test for more than two groups.

```python
from scipy.stats import kruskal

# Example data for three groups
group1 = [1.2, 2.3, 2.1, 3.4, 2.8]
group2 = [2.0, 3.1, 1.8, 2.9, 3.3]
group3 = [1.8, 2.7, 2.5, 3.0, 2.6]

stat, p = kruskal(group1, group2, group3)
print("Kruskal-Wallis H Test:")
print("Statistic:", stat, "p-value:", p)
```

**Friedman Test:**
A non-parametric alternative for repeated measures (more than two related groups).

```python
from scipy.stats import friedmanchisquare

# Example: three repeated measurements for each subject
subject1 = [1.2, 2.3, 2.1]
subject2 = [2.0, 3.1, 1.8]
subject3 = [1.8, 2.7, 2.5]
subject4 = [3.4, 2.9, 3.3]
subject5 = [2.8, 3.0, 2.6]

stat, p = friedmanchisquare(subject1, subject2, subject3, subject4, subject5)
print("Friedman Test:")
print("Statistic:", stat, "p-value:", p)
```

4. Correlation Analysis
Spearman’s Rank Correlation:
Non-parametric measure of rank correlation. It assesses how well the relationship between two variables can be described using a monotonic function.

```python
from scipy.stats import spearmanr

# Example: two sets of measurements
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 6, 5]

corr, p = spearmanr(x, y)
print("Spearman's Rank Correlation:")
print("Correlation:", corr, "p-value:", p)
```

**Summary**

Mann-Whitney U and Wilcoxon Signed-Rank tests are appropriate when comparing two independent or paired groups, respectively.

Kruskal-Wallis and Friedman tests are used for comparing more than two groups, either independent or related.

Spearman’s Rank Correlation assesses the relationship between two variables when the data are not normally distributed.

Each of these tests comes with assumptions and contexts where they are appropriate. Always check your data characteristics (e.g., levels of measurement, sample size) to select the most suitable test for your analysis.
